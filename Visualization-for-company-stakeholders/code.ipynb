```python
#Task 1 - Loan Status
#Importing header files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Code starts here
# Loading the dataset
data = pd.read_csv(path)

# Counting the loan status
loan_status = data['Loan_Status'].value_counts()

# Plotting the bar plot of the loan status
loan_status.plot(kind='bar')
plt.show()

#Code ends here

#Task 2 - Everyone needs money
#Code starts here
# Plotting an unstacked bar plot
property_and_loan = data.groupby(['Property_Area','Loan_Status'])
property_and_loan = property_and_loan.size().unstack()
property_and_loan.plot(kind='bar',stacked=False,figsize=(15,10))

# Changing the labels and titles
plt.xlabel('Property_Area')
plt.ylabel('Loan_Status')
plt.xticks(rotation=45)
plt.show()

#Code ends here

#Task 3 - Expensive Education
#Code starts here
education_and_loan = data.groupby(['Education','Loan_Status'])
education_and_loan = education_and_loan.size().unstack()
education_and_loan.plot(kind='bar', stacked=True, figsize=(15,10))

# Changing the labels and titles
plt.xlabel('Education Status')
plt.ylabel('Loan Status')
plt.xticks(rotation=45)
plt.show()

#Code ends here

#Task 4 - Smarter and Richer? 
#Code starts here
# Subsetting the dataset based on the education level
graduate = data[data['Education'] == 'Graduate']

not_graduate = data[data['Education'] == 'Not Graduate']

# Plotting a density plot for Graduate and Non-Graduate loan borrowers
graduate['LoanAmount'].plot(kind='density',label='Graduate')

not_graduate['LoanAmount'].plot(kind='density',label='Not Graduate')

#Code ends here

#For automatic legend display
plt.legend()

#Task 5 - Income vs. Loan
#Code starts here
# Setting up the subplots
fig, (ax_1, ax_2, ax_3) = plt.subplots(1,3, figsize=(20,8))

# Plotting a scatter plot for Applicant income against Loan amount
ax_1.scatter(data['ApplicantIncome'],data['LoanAmount'])
ax_1.set_title('Applicant Income')

# Plotting a scatter plot for Coapplicant income against Loan amount
ax_2.scatter(data['CoapplicantIncome'],data['LoanAmount'])
ax_2.set_title('Coapplicant Income')

# Creating a new column Total Income
data['TotalIncome'] = data['ApplicantIncome'] + data['CoapplicantIncome']

# Plotting a scatter plot for Total Income against Loan amount
ax_3.scatter(data['TotalIncome'],data['LoanAmount'])
ax_3.set_title('Total Income')

plt.show()

# Code ends here
```
