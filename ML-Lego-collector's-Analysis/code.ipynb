```python
#Task 1 -  Data loading and splitting
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
# code starts here
# Loading the dataset
df = pd.read_csv(path)
print(df.head())

# Feature variables
X = df.drop('list_price', axis=1)

# Target variable
y = df.iloc[:,1]

# Splitting the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=6) 
# code ends here


#Task 2 - Predictor Check!
import matplotlib.pyplot as plt

# code starts here

cols = X_train.columns

# Checking correalation between features using scatter plot
fig, axes = plt.subplots(nrows = 3, ncols = 3, figsize=[25,15])

for i in range(0,3):
    for j in range(0,3):
        col = cols[ i * 3 + j ]
        axes[i,j].scatter(X_train[col],y_train)
        axes[i,j].set_title(col)
        axes[i,j].set_ylabel('Price list of the set')
        axes[i,j].set_xlabel(col)

# code ends here
 

#Task 3 - Reduce feature redundancies!
# Code starts here
# Correlation between the features
corr = X_train.corr()
print(corr)

# dropping columns from X_train
X_train.drop(['play_star_rating','val_star_rating'], axis=1, inplace=True)

# dropping columns from y_train
X_test.drop(['play_star_rating','val_star_rating'], axis=1, inplace=True)

# Code ends here

#Task 4 - Is my price prediction ok?
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Code starts here
# Assigning the regression line
regressor = LinearRegression()

# Fitting the Feature and Target values
regressor.fit(X_train, y_train)

# Predicting the target value for the training dataset
y_pred = np.array(regressor.predict(X_test))
print("Predicted Response :\n", y_pred)

# Metric - Mean squared Error of the model
mse = mean_squared_error(y_test, y_pred)
print("Mean squared Error of the model :", mse)

# Metric - R-squared value of the model
r2 = r2_score(y_test, y_pred)
print("R-squared value of the model:", r2)

# Code ends here

#Task 5 - Residual check
# Code starts here
# Calaculate Residual
residual = y_test - y_pred

# Find the most frequent values of residuals
plt.figure(figsize=[12,5])
residual.hist(bins=25)
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.title("Residual Error plot")
plt.show()

# Code ends here
```
